```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('static/T10Y2Y.csv')
```

```python
df.head()
```

```python
slope = pd.DataFrame({'date': df.date, 't10y2y_slope':np.gradient(df.t10y2y)})
```

```python
slope.head()
```

```python
slope.plot(figsize=(20,5))
```

```python
df.loc[:,'sma'] = df.t10y2y.rolling(356, win_type ='triang').mean()
```

```python
df.plot(figsize=(16,9))
```

```python
slope = pd.DataFrame({'date': df.date, 't10y2y_slope':np.gradient(df.sma)})
```

```python
slope.plot(figsize=(20,5))
```

```python
slope.to_csv('static/T10Y2Y_SMA_SLOPE.csv', index=False)
```
